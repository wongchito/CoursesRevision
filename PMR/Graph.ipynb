```python
%%tikz -f svg -S DAG_serial.svg
\usetikzlibrary{bayesnet}
\tikzstyle{every node} = [latent, fill=none]
\node (z) at (0, 0) {$z$};
\node (x) at +(-1.5, 0) {$x$};
\node (y) at +(1.5, 0) {$y$};
\edge {x} {z};
\edge {z} {y};
```

```python
%%tikz -f svg -S DAG_diverging.svg
\usetikzlibrary{bayesnet}
\tikzstyle{every node} = [latent, fill=none]
\node (z) at (0, 0) {$z$};
\node (x) at +(-1.5, 0) {$x$};
\node (y) at +(1.5, 0) {$y$};
\edge {z} {x,y};
```

```python
%%tikz -f svg -S DAG_converging.svg
\usetikzlibrary{bayesnet}
\tikzstyle{every node} = [latent, fill=none]
\node (z) at (0, 0) {$z$};
\node (x) at +(-1.5, 0) {$x$};
\node (y) at +(1.5, 0) {$y$};
\edge {x,y} {z};
```
